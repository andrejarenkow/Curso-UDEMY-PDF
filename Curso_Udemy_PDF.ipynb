##Leitura das tabelas do arquivo

A primeira que faremos é utilizar a função mais básica do Tabula: o read_pdf, a qual nos retornará uma lista de DataFrames.





```python
tabula.read_pdf(arquivo_pdf)
```

Não é bem o que queríamos, mas já é um início.

Por padrão, se não especificarmos, o read_pdf verifica somente a primeira página do documento. No entanto, podemos utilizar o parâmetro **pages** para colocar aquelas que queremos.

```python
tabula.read_pdf(arquivo_pdf, pages='all')
```

Analisando o arquivo pdf, quase todas as páginas possuem tabelas. Entretanto, as tabelas que nos interessam estão apenas nas páginas 1, 2, 3 e 4.

```python
tabula.read_pdf(arquivo_pdf, pages='1,2,3,4')
```

Também faremos uso do parâmetro lattice, que refere-se a buscar tabelas que tenham linhas separadoras.

```python
tabula.read_pdf(arquivo_pdf, pages='1,2,3,4', lattice=True)
```

A função nos devolve uma lista de DataFrames, porém, desta maneira não conseguimos visualizá-los adequadamente. Sendo assim, vamos criar um loop de **FOR**, imprimindo cada um deles com a função **display** e um pequeno **print** antes de cada um, para sabermos qual é qual.

```python
lista_df = tabula.read_pdf(arquivo_pdf, pages='1,2,3,4', lattice=True)
contador = 0

for df in lista_df:
  print(contador)
  contador = contador + 1
  display(df)
```

Somente queremos os DataFrames que são aquelas com os **ENSAIOS**, ou seja, os Dataframes 3, 5, 6, 8, 9, 10, 12 (lembrando que a contagem começa em 0)

```python
dfs_selecionados = [3,5,6,8,9,10,12]

for i in dfs_selecionados:
  print(i)
  display(lista_df[i])
```

Como queremos juntar todas tabelas em uma só, é fundamental que tenham as mesmas colunas, com os mesmos nomes, para que encaixemos uma abaixo da outra.

```python
lista_dfs_iguais = [5,6,8,9]

for i in lista_dfs_iguais:
  lista_df[i].columns = nome_colunas

lista_df[6]
```

#Arquivo utilizado
Agora vamos "abrir" o arquivo, e deixá-lo armazenado em uma variável chamada **pdf**

```python
pdf = pdfplumber.open('/content/BDI_00_20221228.pdf')
pdf
```

Agora temos que verificar em qual página do nosso arquivo está a informação desejada. No caso, está logo na primeira mesmo, no cabeçalho. Assim, diremos ao pdfplumber guardar a primeira página somente.

```python
primeira_pagina = pdf.pages[0]
primeira_pagina
```

A partir disso, extrairemos o texto contido nesta página com a função **extract_text()** e armazená-lo como a variável **texto**

As informações que queremos são:
* Data - linha 2
* Edição do boletim - linha 2
* Pontos no fechamento da IBOVESPA - linha 3

```python
lista_palavras[3]
```

###Extraindo a data e a edição do boletim

```python
lista_palavras[2].split(' - ')
```

Dentre esta nova lista, o que queremos são o item 1 e 2. Vamos armazená-las cada uma em sua variável.

```python
data = lista_palavras[2].split(' - ')[1]
edicao = lista_palavras[2].split(' - ')[2]
```

Por questões de economia de código, vamos separar apenas no espaço.

**OBS: por padrão, a função split quebra nos espaços**

```python
lista_palavras[3].split()
```

Queremos apenas o quarto valor da lista e vamos armazená-lo em uma variável.

```python
pontos = lista_palavras[3].split()[3]
pontos
```

###Gerando estes dados para mais arquivos

Fazer esta tarefa para um arquivo é relativamente simples, e nem precisaria do Python para realizá-la.

Mas e se temos vários arquivos?

Agora vou mostrar como fazer isso com a biblioteca **glob** e o código que acabamos de criar.

```python
import glob
```

A biblioteca glob é bastante similar à **os**, que é para o manejo de arquivos. No entanto, eu acho ela mais fácil de executar.

O primeiro passo é subir os arquivos que desejamos ler.

Depois, chamamos o glob para criar uma lista com todos os arquivos presentes na pasta e que terminam com a extensão que desejamos, nesse caso, *pdf*.

```python
lista_arquivos = glob.glob('/content/*pdf')
lista_arquivos
```

Agora podemos criar uma função e aplicá-la a cada um dos arquivos e depois transformá-lo em excel ou csv, ou, como é bastante curto, apenas criamos um **FOR**.

Para o armazenamento dos valores, vamos adicionando eles a uma lista vazia (dados) com a função **append**.

Em seguinda, transformamos em um DataFrame com a ajuda do Pandas.

```python
dados = []

for arquivo in lista_arquivos:
  pdf = pdfplumber.open(arquivo)
  primeira_pagina = pdf.pages[0]
  texto = primeira_pagina.extract_text()
  lista_palavras = texto.split('\n')
  data = lista_palavras[2].split(' - ')[1]
  edicao = lista_palavras[2].split(' - ')[2]
  pontos = lista_palavras[3].split()[3]
  dados.append([data,edicao,pontos])

df = pd.DataFrame(dados, columns = ['Data', 'Edição', 'Pontos'])
df
```

### Formatação dos dados
Vale observar que todos os valores presentes no nosso DataFrame são do tipo *string*. Vamos transformar os pontos em números!

Importante ressaltar que, dentro do Python, o separador de milhares não existe e o separador decimal é o ponto! A primeira coisa que devemos fazer então é substituir estes caracteres.

```python
df['Pontos_num'] = df['Pontos'].str.replace('.','',).str.replace(',','.')
df
```

Em seguida, aplicamos a função **pd.to_numeric()** à coluna, e excluímos àquela anterior.

```python
df['Pontos_num'] = pd.to_numeric(df['Pontos_num'])
df.drop('Pontos', axis=1, inplace=True)
df
```

Assim, temos um novo DataFrame, que pode ser exportado ou o que queiramos fazer.
